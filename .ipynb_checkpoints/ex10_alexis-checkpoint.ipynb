```python
from collections import Counter
import csv
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np

#calculate/make list of top 15 senders and make graphs 
hillary_emails_csv = open("H_Clinton-emails.csv")
email = csv.reader(hillary_emails_csv)
senders = []

for row in email:
    senders.append(row[4].strip())
top_senders = Counter(senders).most_common(15)
print ("\nThe Top 15 Senders Are:\n")
for pair in top_senders:
    print (pair[0])
 
names = (top_senders)
y_pos = np.arange(len(names))
occurence = [val[1] for val in top_senders]
plt.bar(y_pos, occurence, align='center', alpha=0.5)
plt.xticks(y_pos, names)
plt.ylabel('Emails Received')
plt.title("Hillary Clinton's Top senders of emails")
plt.show()
```

```python

#create list of contents 
#create wordcloud of top words  
hillary_emails_csv = open("H_Clinton-emails.csv")
email = csv.reader(hillary_emails_csv)
word_list = []
for row in email:
    word_list.append(row[21])

words = []
stpwords = []
stopwords_file = open("stopwords_en.txt")
for word in stopwords_file:
    stpwords.append( word.strip() )
for word_list, words in [(word_list, words)]:
    for line in word_list:
        parts = line.strip().split()
        for word in parts:
            if word.lower() not in stpwords:
                words.append(word.lower())
content = [word.lower() for word in words if word.isalpha()]
text_str = ' '.join(content)
```

```python
# Defining the wordcloud parameters
wc = WordCloud(background_color="white", max_words=500,
               stopwords=stpwords)


# Generate word cloud for top words
wc.generate(text_str)


# Store to file
wc.to_file('Top Words.png')

# Show the cloud
plt.imshow(wc)
plt.axis('off')
plt.show()
```
