```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
iris=datasets.load_iris()
print(type(iris))
```

```python
import pandas as pd
iris=pd.DataFrame(data=np.c_[iris['data'],iris['target']],columns=iris['feature_names']+['species'])
iris.head(4)
```

```python
iris.species = np.where(iris.species == 0.0, 'setosa', np.where(iris.species == 1.0,'versicolor','virginica'))
```

```python
iris.hist()
plt.show()
```

```python
iris['sepal length (cm)'].hist()
plt.show()
```

```python
plt.hist(iris['sepal length (cm)'], bins=10, histtype='step')
plt.show()
```

```python
plt.hist(iris['sepal length (cm)'], alpha=.3)
sns.rugplot(iris['sepal length (cm)']);
plt.show()
```

```python
g = sns.FacetGrid(iris, col="species", margin_titles=True) #frequency disribution in terms of sex and time
#bins = np.linspace(0, 60, 13)
g.map(plt.hist, "sepal length (cm)", color="steelblue", lw=0)
plt.show()
```

```python
iris.boxplot()
plt.title("iris")
plt.show()
```

```python
iris.boxplot(column="sepal length (cm)",by="species")
plt.title("Iris")
plt.show()
```

```python
sns.boxplot(x="sepal length (cm)", y="sepal width (cm)", hue="species", data=iris);
plt.show()
```

```python
plt.scatter(iris['sepal length (cm)'],iris['sepal width (cm)'],
            # the marker as
            marker='x',
            # the color
            color='b',
            # the alpha
            alpha=0.7,
            # with size
            s = 124)

plt.title("length vs width")
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.show()
```

```python
iris.plot(kind="scatter", x="sepal length (cm)", y="sepal width (cm)")
plt.show()
```

```python
sns.FacetGrid(iris, hue="species", height=5).map(plt.scatter, "sepal length (cm)", "sepal width (cm)").add_legend()
plt.show()
```

```python
iris['sepal length (cm)'].plot.bar()
plt.show()
```

```python
iris.set_index('sepal length (cm)').plot.bar(rot=0, title='Rankings', figsize=(15,10), fontsize=12)
plt.show()
```

```python
sd=iris.groupby(['species']).std() 
isa=iris.groupby(['species']).mean()
```

```python
my_plot = isa.plot(kind='bar') #barplot of the average meaurement values for the 3 species
my_plot.set_ylabel("Average values")
plt.show()
```

```python
isa.plot(kind='bar', yerr=sd)
plt.show()
```
